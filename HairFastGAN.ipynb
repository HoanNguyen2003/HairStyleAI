```python
# Display function
def display_images(images=None, **kwargs):
    is_titles = images is None
    images = images or kwargs

    grid = gridspec.GridSpec(1, len(images))
    fig = plt.figure(figsize=(20, 10))

    for i, item in enumerate(images.items() if is_titles else images):
        title, img = item if is_titles else (None, item)

        img = T.functional.to_pil_image(img) if isinstance(img, torch.Tensor) else img
        img = Image.open(img) if isinstance(img, str | Path) else img

        ax = fig.add_subplot(1, len(images), i+1)
        ax.imshow(img)
        if title:
            ax.set_title(title, fontsize=20)
        ax.axis('off')

    plt.show()
```

```python
import cv2
import numpy as np
import mediapipe as mp
import json
from datetime import datetime

def get_face_shape(image_path):
    mp_face_mesh = mp.solutions.face_mesh
    img = cv2.imread(str(image_path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    with mp_face_mesh.FaceMesh(static_image_mode=True) as face_mesh:
        results = face_mesh.process(img_rgb)
        if not results.multi_face_landmarks:
            return "unknown"
        landmarks = results.multi_face_landmarks[0]
        xs = [lm.x for lm in landmarks.landmark]
        ys = [lm.y for lm in landmarks.landmark]
        width = max(xs) - min(xs)
        height = max(ys) - min(ys)
        ratio = width / height
        if ratio > 0.9:
            return "round"
        elif ratio < 0.75:
            return "long"
        else:
            return "oval"

def get_hair_color(image_path):
    img = cv2.imread(str(image_path))
    h, w, _ = img.shape
    hair_region = img[int(h*0.1):int(h*0.3), int(w*0.3):int(w*0.7)]
    avg_color = hair_region.mean(axis=(0,1))
    hex_color = '#%02x%02x%02x' % tuple(avg_color.astype(int)[::-1])
    return hex_color

def get_color_family(hex_color):
    """Đơn giản: chỉ classify màu cơ bản"""
    hex_color = hex_color.lstrip('#')
    r = int(hex_color[0:2], 16)
    g = int(hex_color[2:4], 16)
    b = int(hex_color[4:6], 16)

    if r > 200 and g > 150 and b < 120:
        return "blonde"
    elif r < 80 and g < 60 and b < 50:
        return "black"
    elif r > 150 and r > g * 1.2:
        return "red"
    else:
        return "brown"

def get_color_name(hex_color):
    """Map hex sang tên màu đơn giản"""
    color_family = get_color_family(hex_color)
    color_names = {
        "blonde": "Golden Blonde",
        "black": "Deep Black",
        "red": "Auburn Red",
        "brown": "Chestnut Brown"
    }
    return color_names.get(color_family, "Natural Brown")

def get_face_advice(face_shape):
    """Lời khuyên đơn giản theo face shape"""
    advice = {
        "oval": {
            "suitable": ["bob", "layers", "bangs", "curls"],
            "avoid": ["very-short"],
            "tip": "Lucky you! Oval faces suit most hairstyles."
        },
        "round": {
            "suitable": ["long-layers", "side-part", "height-styles"],
            "avoid": ["chin-length", "center-part"],
            "tip": "Add length and avoid width at cheek level."
        },
        "long": {
            "suitable": ["bangs", "width-styles", "waves"],
            "avoid": ["very-long", "center-part"],
            "tip": "Add width and avoid extra length."
        }
    }
    return advice.get(face_shape, advice["oval"])

def get_compatibility_score(face_shape):
    """Điểm tương thích đơn giản"""
    scores = {
        "oval": 10,    # Perfect for most styles
        "round": 7,    # Good but need some care
        "long": 6      # Need specific styles
    }
    return scores.get(face_shape, 8)

def generate_simple_semantic(image_path, existing_metadata=None):
    """Tạo semantic data ĐƠN GIẢN"""

    # 1. Get basic data
    face_shape = get_face_shape(image_path)
    hair_color = get_hair_color(image_path)
    color_family = get_color_family(hair_color)
    color_name = get_color_name(hair_color)

    # 2. Get advice
    face_advice = get_face_advice(face_shape)
    compatibility_score = get_compatibility_score(face_shape)

    # 3. Generate simple semantic data
    semantic_data = {
        "name": f"{face_shape.title()} Face {color_name} Style",
        "description": f"AI hairstyle for {face_shape} face with {color_name} color. {face_advice['tip']}",
        "image": existing_metadata.get('image', '') if existing_metadata else '',

        # Simple semantic info
        "faceShape": face_shape,
        "hairColor": hair_color,
        "colorFamily": color_family,
        "colorName": color_name,
        "compatibilityScore": compatibility_score,
        "advice": face_advice,

        # Simple attributes
        "attributes": [
            {"trait_type": "Face Shape", "value": face_shape},
            {"trait_type": "Hair Color", "value": hair_color},
            {"trait_type": "Color Family", "value": color_family},
            {"trait_type": "Color Name", "value": color_name},
            {"trait_type": "Compatibility", "value": f"{compatibility_score}/10"},
            {"trait_type": "Suitable Styles", "value": ", ".join(face_advice['suitable'])},
            {"trait_type": "Avoid Styles", "value": ", ".join(face_advice['avoid'])},
            {"trait_type": "AI Tip", "value": face_advice['tip']},
            {"trait_type": "Analysis Date", "value": datetime.now().strftime("%Y-%m-%d")},
            {"trait_type": "Platform", "value": "HairStyleAI"}
        ]
    }

    return semantic_data
```

```python
import io
import nest_asyncio
from fastapi import FastAPI, File, UploadFile
from fastapi.responses import StreamingResponse
from hair_swap import HairFast, get_parser
import torchvision.transforms.functional as TF
from pyngrok import ngrok
from pydantic import BaseModel
import os
import base64
from fastapi.responses import JSONResponse
import google.generativeai as genai

import uvicorn

# --- Khởi model ---
args = get_parser().parse_args([])    # parse args rỗng như bạn đang làm
hair_fast = HairFast(args)

# --- Tạo app ---
app = FastAPI()

# Cấu hình API Key
os.environ["GOOGLE_API_KEY"] = "AIzaSyB9HpdiuoHzYSaAiRYL5gVSHZJy9M6WWOg"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# Khởi tạo mô hình
model = genai.GenerativeModel("gemini-1.5-pro-latest")

# Khởi tạo ứng dụng FastAPI
app = FastAPI()

# Định nghĩa dữ liệu đầu vào
class ChatRequest(BaseModel):
    prompt: str
    max_tokens: int = 100

# Định nghĩa endpoint
@app.post("/chat")
async def chat(request: ChatRequest):
    response = model.generate_content(
        request.prompt,
        generation_config={"max_output_tokens": request.max_tokens}
    )
    return {"response": response.text}


@app.post("/swap/")
async def swap(
    face: UploadFile = File(...),
    shape: UploadFile = File(...),
    color: UploadFile = File(...)
):
    print("Received files:", face.filename, shape.filename, color.filename)
    b_face = await face.read()
    b_shape = await shape.read()
    b_color = await color.read()

    from tempfile import NamedTemporaryFile
    tmpf = NamedTemporaryFile(suffix=".jpg", delete=False); tmpf.write(b_face); tmpf.flush()
    tmps = NamedTemporaryFile(suffix=".png", delete=False); tmps.write(b_shape); tmps.flush()
    tmpc = NamedTemporaryFile(suffix=".png", delete=False); tmpc.write(b_color); tmpc.flush()

    # gọi model
    out_tensor, *_ = hair_fast.swap(tmpf.name, tmps.name, tmpc.name, align=True)
    pil = TF.to_pil_image(out_tensor)
    pil.save("result.png")

    # --- PHÂN TÍCH SEMANTIC từ ảnh gốc (face) ---
    face_shape = get_face_shape(tmpf.name)  # Phân tích từ ảnh face gốc
    hair_color_hex = get_hair_color(tmpc.name)  # Lấy màu tóc từ kết quả
    color_family = get_color_family(hair_color_hex)         # Nhóm màu
    color_name = get_color_name(hair_color_hex)             # Tên màu
    face_advice = get_face_advice(face_shape)               # Gợi ý
    compatibility_score = get_compatibility_score(face_shape)  # Điểm

    # --- Sinh semantic metadata ---
    semantic = generate_simple_semantic(tmpf.name, existing_metadata={
        "image": "result.png"
    })

    # --- Encode ảnh kết quả dưới dạng base64 ---
    buf = io.BytesIO()
    pil.save(buf, format="PNG")
    buf.seek(0)
    img_base64 = base64.b64encode(buf.read()).decode("utf-8")

    return JSONResponse({
        "image_base64": img_base64,
        "semantic": semantic
    })

# --- Bật ngrok tunnel (để máy local gọi được) ---
nest_asyncio.apply()
public_url = ngrok.connect(8000, bind_tls=True)
print(" → Public URL:", public_url)

# --- Chạy server trong cell này luôn ---
uvicorn.run(app, host="0.0.0.0", port=8000)
```
